```python
import nltk
import requests
import random
import urllib.request
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from bs4 import BeautifulSoup
from nltk.corpus import stopwords # Palabras de parada 
```

```python
response = urllib.request.urlopen("https://poemario.org/20-poemas-amor-una-cancion-desesperada/")
html = response.read()
print(html)
```

```python
#Limpia etiquetas HTML5
soup = BeautifulSoup(html, "html5lib")
text = soup.get_text(strip=True)
print(text)
```

```python
#Separamos texto por tokens (Se repiten)
tokens = [t for t in text.split()]
print(tokens)
```

```python
# Frecuencia con la que se repiten los tokens 
frecuencia = nltk.FreqDist(tokens)
for key,val in frecuencia.items():
    print(str(key) + ":" + str(val))
#print(frecuencia)
```

```python
frecuencia.plot(20, cumulative=False)
```
